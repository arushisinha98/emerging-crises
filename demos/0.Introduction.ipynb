## Developed Markets
The frequency and duration of financial crises is studied to understand the stability of economies and to inform investments in financial instruments such as government bonds and real estate. The study of financial crises is biased towards developed markets, for which we have rich historical information and data sets. Developed markets often show patterns of crisis concurrence which are periods where multiple markets are in crisis. This is often due to price shocks, credit crises, or asset bubbles introducing stress to the global financial market and having spillover effects in other economies.

Financial crises are also rare events, especially in developed markets. Rather than predict when they start, it is often more interesting to predict when they might be imminent (i.e. 1-2 years prior). This is often a stable economic period so it is interesting to understand if there are undercurrents of crisis precursors that exist that can separate these pre-crisis period from other economic states.

The plot below shows the 17 most studied developed markets since 1800. Each country's positive crisis labels (i.e. 2 years prior to a financial crisis) are plotted as colored bars where the colors group countries by continent. Grey shaded regions denote years during which at least five countries concurrently have positive crisis labels (i.e. at least five countries are due to experience a financial crisis).

## So What?

### 1. Bond Spread Compression Mispricing
Bond "spreads" are the extra interest rates that risky countries pay compared to their developed market counterparts to entice investment. When these spreads are too low, investors are not charging enough for the risk they're taking.
- Crisis prediction models rely on economic data to estimate risk.
- Emerging markets often have poor quality, delayed, or missing data.
- Models based on incomplete data underestimate the true risk.
- Investors see the low risk estimates and demand lower interest rates.
- Emerging market governments can borrow money cheaply.

This creates a dangerous cycle where cheap borrowing encourages governments to take on more debt than they should, making a future crisis more likely.

### 2. "Wake Up Call" Effects
When a major crisis event occurs, everyone tries to sell emerging market bonds at once. Interest rates spike dramatically and quickly. Even countries that are not in crisis get caught up in the panic selling, making it expensive for all emerging markets to borrow money. This can be reflected in emerging market equity ETFs, even if they have minimal direct exposure to the market initially in crisis.

### 3. Central Bank Interventions
When foreign investors sell emerging market bonds rapidly, the country's central bank steps in to buy them.
- Bond prices fall and interest rates rise sharply.
- The country's central bank starts buying its own government's bonds to stabilize prices.

In the short-term, this stops the panic and stabilizes markets. In the long-term, the central bank is essentially printing money to fund government spending, which can lead to inflation.

### 4. Policy Trade-offs
When borrowing looks cheap and safe, governments make decisions that create hidden future risks.
- Low interest rates make borrowing attractive.
- Governments issue longer-term bonds.
- They might borrow in foreign currencies because rates are even lower. 

However, longer-term debt means more vulnerability to changing market conditions. Foreign currency debt becomes much more expensive if the local currency weakens. When crisis warning signs appear, it becomes much harder and more expensive to refinance this debt.
